```python
#Primero voy a crear una lista con las urls de los barrios
response=requests.get("https://www.indomio.es/venta-casas/madrid-capital/")

soup = BeautifulSoup(response.content)
# Encontrar todos los enlaces con la clase 'in-breadcrumbLink__dropdownLink'
links = soup.find_all("a", class_="in-breadcrumbLink__dropdownLink")

# Usar un conjunto para almacenar URLs únicas
#unique_hrefs = set()

# Usar una lista para almacenar URLs únicas
curl_barrios = []

for clink in links:
    href = clink.get('href')
    if href and href not in curl_barrios:  # Asegurarse de que el href no sea None y no esté ya en la lista
        curl_barrios.append(href)


# Mostrar la lista completa
print(curl_barrios)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, WebDriverException
from bs4 import BeautifulSoup
import pandas as pd
import math
import time
x=0
# Configuración del driver
path_driver_edge = 'C:/Mas/Data_Analyst/Entregas/msedgedriver.exe'
options = webdriver.EdgeOptions()
options.add_argument("--start-maximized")
service = webdriver.EdgeService(executable_path=path_driver_edge)
driver = webdriver.Edge(service=service, options=options)
arefyacargadas=[]

# Variables de almacenamiento
data = {
    "precio": [],
    "title": [],
    "url": [],
    "atributos": [],
    "atributos2": [],
    "read_all": [],
    "badges": [],
    "consumoenergia": [],
    "emisiones_co2": [],
    "carc_todas": []
}

# Recorrer URLs
for curlbar in curl_barrios:
   
    driver.get(curlbar)
    WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CLASS_NAME, 'in-realEstateListHeader__title')))
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Número de páginas
    num_text = (soup.find('div', class_='in-realEstateListHeader__title').text[:3])
    num = int(''.join(filter(str.isdigit, num_text)))
    nfinpagina = math.ceil(num / 22)
   
    
    for npag in range(1, nfinpagina + 1):
        curl = f"{curlbar}?pag={npag}" if npag > 1 else curlbar
        driver.get(curl)
        try:
            WebDriverWait(driver, 15).until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'nd-mediaObject__content')))
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            
            pisos = soup.find_all('div', class_='nd-mediaObject__content in-listingCardPropertyContent')
            for piso in pisos:
                href = piso.find('a', class_='in-listingCardTitle')['href']
                if href in list_url:
                    arefyacargadas.append(href)
                else:
                    x=x+1
                    driver.get(href)
                    try:
                        WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CLASS_NAME, 're-title__title')))
                        soup = BeautifulSoup(driver.page_source, 'html.parser')

                        # Extracción de datos
                        title = soup.find('h1', class_='re-title__title').get_text(strip=True) if soup.find('h1', class_='re-title__title') else "Título no encontrado"
                        price = soup.find('div', class_='re-overview__price').get_text(strip=True) if soup.find('div', class_='re-overview__price') else "Precio no encontrado"
                        badges = [badge.get_text(strip=True) for badge in soup.find_all('li', class_='re-featuresBadges__badge')]
                        characteristics = [feature.span.get_text(strip=True) for feature in soup.find_all('div', class_='re-mainFeatures__item')]
                        listacaract = "|".join(characteristics)
                        
                        listacontenido = "|".join([
                            f"{content.find('dt').get_text(strip=True)}### {content.find('dd').get_text(strip=True)}"
                            for content in soup.find_all('div', class_='re-featuresItem')
                        ])
                        
                        consumo_energia, emisiones_co2 = "", ""
                        main_consumptions = soup.find('div', class_='re-mainConsumptions')
                        if main_consumptions:
                            consumo_energia_element = main_consumptions.find('p', string='Consumo de energia')
                            if consumo_energia_element:
                                next_p = consumo_energia_element.find_next('p')
                                consumo_energia = next_p.get_text(strip=True) if next_p else ""
                            
                            emisiones_co2_element = main_consumptions.find('p', string='Emisiones de CO₂')
                            if emisiones_co2_element:
                                next_p = emisiones_co2_element.find_next('p')
                                emisiones_co2 = next_p.get_text(strip=True) if next_p else ""
                        
                        
                        
                        # Desplazarse hasta el botón antes de hacer clic utilizando JavaScript
                        boton_abrir = WebDriverWait(driver, 10).until(
                            EC.element_to_be_clickable((By.CSS_SELECTOR, ".nd-button.re-primaryFeatures__openDialogButton"))
                        )
                        driver.execute_script("arguments[0].scrollIntoView(true);", boton_abrir)
                        time.sleep(1)  # Esperar un momento para asegurarse de que el desplazamiento se haya completado
                        driver.execute_script("arguments[0].click();", boton_abrir)

                        # Esperar a que la ventana emergente se cargue
                        time.sleep(5)  # Ajustar según la carga dinámica

                        # Obtener el contenido de la página
                        soup = BeautifulSoup(driver.page_source, 'html.parser')

                        # Encontrar todos los elementos con la clase nd-dialogFrame__content
                        dialog_frame_contents = soup.find_all('div', class_='nd-dialogFrame__content')
                        listatodas = ""
                        for frame in dialog_frame_contents:
                            features = frame.find_all('div', class_='re-primaryFeaturesDialogSection__feature')
                            for feature in features:
                                title3 = feature.find('dt', class_='re-primaryFeaturesDialogSection__featureTitle').text
                                description = feature.find('dd', class_='re-primaryFeaturesDialogSection__featureDescription').text
                                listatodas += f"|{title3}### {description}"

                        read_all = soup.find('div', class_='in-readAll').text if soup.find('div', class_='in-readAll') else "Leer todo no encontrado"
                        
                        # Agregar datos al diccionario
                        # Agregar datos al diccionario
                        data["precio"].append(price)
                        data["title"].append(title)
                        data["url"].append(href)
                        data["atributos"].append(listacaract)
                        data["atributos2"].append(listacontenido)
                        data["badges"].append(badges)
                        data["consumoenergia"].append(consumo_energia)
                        data["emisiones_co2"].append(emisiones_co2)
                        data["read_all"].append(read_all)
                        data["carc_todas"].append(listatodas)
                        #data["url"].append(href)
                    
                    except TimeoutException:
                        print("No se encontró la página de detalles, continuando con el siguiente piso...")
                        # Agregar valores predeterminados para mantener la longitud de las listas
                        data["carc_todas"].append("N/A")
                        data["read_all"].append("N/A")                       
                        data["precio"].append(price)
                        data["title"].append(title)
                        data["url"].append(href)
                        data["atributos"].append("N/A")
                        data["atributos2"].append("N/A")
                        data["badges"].append("N/A")
                        data["consumoenergia"].append("N/A")
                        data["emisiones_co2"].append("N/A")
                
                    except WebDriverException:
                        print("Error de WebDriver, continuando con el siguiente piso...")
                        # Agregar valores predeterminados para mantener la longitud de las listas
                        data["carc_todas"].append("N/A")
                        data["read_all"].append("N/A")                        
                        data["precio"].append(price)
                        data["title"].append(title)
                        data["url"].append(href)
                        data["atributos"].append("N/A")
                        data["atributos2"].append("N/A")
                        data["badges"].append("N/A")
                        data["consumoenergia"].append("N/A")
                        data["emisiones_co2"].append("N/A")

                print(f" nº pisos {x}")
            
                
        except TimeoutException:
            print("No se encontró la página, continuando con la siguiente...")
        except WebDriverException:
            print("Error de WebDriver, continuando con la siguiente página...")

# Crear el DataFrame
    df = pd.DataFrame(data)
    df.to_csv('scraping_indomio06012025.csv', index=False, encoding='utf-8')
# Crear el DataFrame
df = pd.DataFrame(data)
df.to_csv('scraping_indomioEnero2025.csv', index=False, encoding='utf-8')

# Cerrar el driver
driver.quit()
```
